```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
input_text = "Hello Nupur Pusha here ."
inputs = tokenizer(input_text, return_tensors="pt")
```

```python
outputs = model(**inputs)
predictions = torch.argmax(outputs.logits, dim=1)
```

```python
predicted_label = tokenizer.decode(predictions[0])
print(f"Predicted label: {predicted_label}")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
input_text = "The weather today is"
input_ids = tokenizer.encode(input_text, return_tensors="pt")
```

```python
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(f"Generated text: {generated_text}")
```

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
```

```python
model_name = "bert-large-uncased-whole-word-masking-finetuned-squad"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
```

```python
context = "The Apollo program, also known as Project Apollo, was the third United States human spaceflight program carried out by NASA."
question = "What was the Apollo program?"
inputs = tokenizer(question, context, return_tensors="pt")
```

```python
outputs = model(**inputs)
```

```python
start_scores = outputs.start_logits
end_scores = outputs.end_logits
```

```python
start_index = torch.argmax(start_scores)
end_index = torch.argmax(end_scores)
answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][start_index:end_index+1]))
```

```python
print(f"Question: {question}")
print(f"Answer: {answer}")
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
import torch
```

```python
model_name = "dbmdz/bert-large-cased-finetuned-conll03-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)
```

```python
input_text = "Apple Inc. is an American multinational technology company headquartered in Cupertino, California."
inputs = tokenizer(input_text, return_tensors="pt")
```

```python
outputs = model(**inputs)
predictions = torch.argmax(outputs.logits, dim=2)
```

```python
tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])
entities = []
current_entity = None

for token, label_idx in zip(tokens, predictions[0]):
    label = tokenizer.convert_ids_to_tokens([label_idx.item()])[0]

    if token.startswith("##"):
        current_entity = (current_entity[0] + token[2:], current_entity[1])
    else:
        if current_entity:
            entities.append(current_entity)
        current_entity = (token, label)
if current_entity:
    entities.append(current_entity)
```

```python
print(f"Input text: {input_text}")
print("Named Entities:")
for entity in entities:
    print(f"Entity: {entity[0]}, Type: {entity[1]}")
```

```python
from transformers import AutoFeatureExtractor, AutoModelForImageClassification
from PIL import Image
import requests
```

```python
feature_extractor = AutoFeatureExtractor.from_pretrained("google/vit-base-patch16-224-in21k")
model = AutoModelForImageClassification.from_pretrained("google/vit-base-patch16-224-in21k")
```

```python
import pandas as pd
import numpy as np
import requests
from PIL import Image
import io
image_url="https://buffer.com/library/content/images/size/w1200/2023/10/free-images.jpg"
response = requests.get(image_url)
image = Image.open(io.BytesIO(response.content))
```

```python
inputs = feature_extractor(images=image, return_tensors="pt")
```

```python
outputs = model(**inputs)
```

```python
predicted_class = outputs.logits.argmax(dim=1)
print(f"Predicted class index: {predicted_class.item()}")
```
